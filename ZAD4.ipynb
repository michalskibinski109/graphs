```python
def find_cluster(node, clusters):
    for cluster in clusters:
        if node in cluster:
            return cluster
def update_clusters(clusters, a, b):
    new_cluster = a.union(b)
    clusters.remove(a)
    clusters.remove(b)
    # print("removed", a, b)
    # print('new', new_cluster)
    clusters.append(new_cluster)
```

```python
spanning_tree = nx.Graph()
edges_queue = sorted(G.edges(data=True), key=lambda x: x[2]['weight'])
clusters = [set([x]) for x in G.nodes]
while edges_queue:
    u, v, w = edges_queue.pop(0)
    claster_u = find_cluster(spanning_tree, u, clusters)
    claster_v = find_cluster(spanning_tree, v, clusters)
    if claster_u == claster_v:
        continue
    else:
        spanning_tree.add_edge(u, v, **w)
        update_clusters(clusters, claster_u, claster_v)
    nx.draw_networkx(spanning_tree)
    plt.show()
```

# Prime

```python
def prims_mst(graph):
    postitive_inf = float('inf')
    selected_nodes = [False for node in range(len(graph.nodes))]
    MST = nx.Graph()

    while(False in selected_nodes):
        minimum = postitive_inf
        start = 0
        end = 0
        for i in range(len(graph.nodes)):
            if selected_nodes[i]:
                for j in range(len(graph.nodes)):
                    if (not selected_nodes[j] and graph[i][j]['weight']>0):  
                        if graph[i][j]['weight'] < minimum:
                            minimum = graph[i][j]['weight']
                            start, end = i, j
    
        selected_nodes[end] = True
        
        if minimum != postitive_inf:
            MST.add_edge(start, end, weight=minimum)
    return MST
```

```python
g = prims_mst(G)
```
